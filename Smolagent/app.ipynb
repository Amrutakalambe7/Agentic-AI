```python
model = LiteLLMModel(
    model_id= "gemini/gemini-1.5-flash",
    api_key= os.getenv("GOOGLE_API_KEY")
)
```

```python
ml_code_agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    additional_authorized_imports=["pands", "numpy", "sklearn","jason"],
    model=model
    )
```

```python
def genearte_ml_code(prompt):
    full_prompt ="""
    You are an AI assistant helping with machine learning code. 
    Complete below 9 steps in a structured manner.

    1. **Load the Diabetes dataset** using sklearn.
    2. **Split the data** (75% train / 25% test).
    3. **Train a linear regression model**.
    4. **Find the best 3-feature combination** (lowest RMSE) for the model.
    5. ** find the R2 & adjusted R2** for the model.
    6. **Apply feature engineering** if needed.
    7. **List the top 5 models based on RMSE**.
    8. **Output the results in a structured table**.
    9. **Plot the result**.

    Ensure that you return:
    - The Python code for each step.
    - The RMSE values of the models.
    - The final selected feature set.
    """
    try:
        response = ml_code_agent.run(full_prompt)
        return response
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
iface = gr.Interface(
    fn=genearte_ml_code, 
    inputs="text", 
    outputs="text",
    title="ML Code Generator using Gemini 1.5 Flash Model",
    description ="Enter a machine learing codeing task and AI will generate Python code for you.")
```

```python
iface.launch()
```
